```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
start = '2015-01-01'
end = '2025-10-25'
stock = 'CODX'

data = yf.download(stock, start, end)
```

```python
data
```

```python
data.reset_index(inplace=True)
```

```python
data
```

```python
#moving average for 100 days
ma_100_days = data.Close.rolling(100).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r')
plt.plot(data.Close, 'g')
plt.show()
```

```python
#moving averages for 200 days
ma_200_days = data.Close.rolling(200).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days,'r')
plt.plot(ma_200_days,'b')
plt.plot(data.Close,'g')
plt.show()
```

```python
data.dropna(inplace=True)
```

```python
#splitting the data into 80% for training and 20% for testing
data_train = pd.DataFrame(data.Close[0: int(len(data)*0.80)])
data_test = pd.DataFrame(data.Close[int(len(data)*0.80): len(data)])
```

```python
data_train.shape[0]
```

```python
data_test.shape[0]
```

```python
#scaling , fit the stock price with 0 & 1 instead of actual stock price
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_train_scale = Scaler.fit_transform(data_train)
```

```python
#array slicing
x=[]
y=[]
for i in range(100, data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
```

```python
x,y = np.array(x) , np.array(y)
```

```python
#train the model 
#keras - comes under tensorflow
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model=Sequential()
model.add(LSTM(units=50,activation='relu',return_sequences=True,input_shape=((x.shape[1],x.shape[2]))))
model.add(Dropout(0.2))

model.add(LSTM(units=60,activation='relu',return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(units=80,activation='relu',return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(units=120,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(x,y, epochs=50, batch_size=32, verbose=1)
```

```python
model.summary()
```

```python
pas_100_days=data_train.tail(100)
```

```python
data_test=pd.concat([pas_100_days, data_test], ignore_index=True)
```

```python
#after adding last 100 days train data set
data_test
```

```python
#need to scale that test data
data_test_scale = Scaler.fit_transform(data_test)
```

```python
#array slicing
x=[]
y=[]
for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])

x,y=np.array(x),np.array(y)
```

```python
y_predict=model.predict(x)
```

```python
#values will be fitted between 0 & 1
y_predict
```

```python
#change that to original stock value
scale=1/Scaler.scale_
```

```python
#predicted value
y_predict = y_predict*scale
```

```python
#actual value
y=y*scale
```

```python
plt.figure(figsize=(10,8))
plt.plot(y_predict,'r',label='Predicted Price')
plt.plot(y,'g',label='Original Price')
plt.xlabel('Time')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
model.save('Stock Prediction Model.keras')
```
